# 创建节点

在NetworkX中创建单个节点、创建多个节点、图本身作为节点

```python
import networkx as nx

import matplotlib.pyplot as plt
%matplotlib inline

plt.rcParams['font.sans-serif'] = ['Microsoft YaHei']
plt.rcParams['axes.unicode_minus']=False
```

创建无节点、无连接的空图

```python
G = nx.Graph()
G
```

```python
G.nodes
```

```python
# 可视化
nx.draw(G)
```

添加单个节点

```python
G.add_node('刘备')
G.nodes
```

```python
G.add_node('Tommy')
G.nodes
```

```python
G.add_node(1)
G.nodes
```

添加多个节点

```python
G.add_nodes_from(['诸葛亮', '曹操'])
G.nodes
```

```python
G.add_nodes_from(range(100, 105))
G.nodes
```

添加带属性特征的节点

```python
G.add_nodes_from([
    ('关羽', {'武器': '青龙偃月刀', '武力值': 90, '智力值': 80}),
    ('张飞', {'武器': '丈八蛇矛', '武力值': 85, '智力值': 75}),
    ('吕布', {'武器': '方天画戟', '武力值': 100, '智力值': 70}),
])
G.nodes
```

```python
# 可视化
nx.draw(G)
```

创建另一个首尾相连的Path Graph

```python
H = nx.path_graph(10)

# 可视化
nx.draw(H)
```

```python
H.nodes
```

将H的节点添加到G中

```python
G.add_nodes_from(H)
G.nodes
```

```python
len(G)
```

```python
nx.draw(G)
```
