### 将图像转换成结构化数组和标量

```python
# 1. 下载图像样本
!wget https://www.dropbox.com/s/l98leemr7r5stnm/Hemanvi.jpeg
```

```python
# 2. 导入cv2和matplotlib 将下载的图像读入到Python环境中
%matplotlib inline
import cv2, matplotlib.pyplot as plt
img = cv2.imread('Hemanvi.jpeg')
```

```python
# 3. 在图像的50~250行和第40~240列之间进行裁剪, 最后使用下列代码将图像转换为灰度图像并绘制
# Crop Image
img = img[50:250,40:240]
# Convert Image to grayscale
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
# Show Image
plt.imshow(img_gray, cmap='gray')
```

```python
# 4. 将图像转换为一个25 x 25的数组, 并绘制出来
img_gray_small = cv2.resize(img_gray,(25,25))
plt.imshow(img_gray_small, cmap='gray')
```

```python
# 5. 检查像素值
print(img_gray_small)
```
