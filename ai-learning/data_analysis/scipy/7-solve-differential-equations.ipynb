### 积分求解

### 简单的例子

$\frac{\text{d}y}{\text{d}t}=sin(t)$


```python
def dy_dt(y, t):
    return sin(t)
```

积分求解：

```python
from scipy.integrate import odeint

t = np.linspace(0, 2 * pi, 100)

result = odeint(dy_dt, 0, t)
```

```python
fig = figure(figsize=(12,4))
p = plot(t, result, "rx", label=r"$\int_{0}^{x}sin(t) dt $")
p = plot(t, -cos(t) + cos(0), label=r"$cos(0) - cos(t)$")
p = plot(t, dy_dt(0, t), "g-", label=r"$\frac{dy}{dt}(t)$")
l = legend(loc="upper right")
xl = xlabel("t")
```

### 高阶微分方程

抛物运动（竖直方向）：

$\frac{d^{2}x}{d^{2}t}=g-\frac{D}{m}\frac{\text{d}x}{\text{d}t}$

改写成如下形式:

$
y=
\begin{bmatrix}
{y},{\frac{\text{d}x}{\text{d}t}}\\
\end{bmatrix}
$

$
\frac{\text{d}y_0}{\text{d}t}=y_1
$

$
\frac{\text{d}y_1}{\text{d}t}=-g-\frac{D}{m}y_1
$

```python
def dy_dt(y, t):
    """Governing equations for projectile motion with drag.
 y[0] = position
 y[1] = velocity
 g = gravity (m/s2)
 D = drag (1/s) = force/velocity
 m = mass (kg)
 """
    g = -9.8
    D = 0.1
    m = 0.15
    dy1 = g - (D/m) * y[1]
    dy0 = y[1] if y[0] >= 0 else 0.
    return [dy0, dy1]
```

```python
position_0 = 0.
velocity_0 = 100
t = linspace(0, 12, 100)
y = odeint(dy_dt, [position_0, velocity_0], t)
```

```python
p = plot(t, y[:,0])
yl = ylabel("Height (m)")
xl = xlabel("Time (s)")
```
