## PageRank 节点重要度

```python
import networkx as nx
import matplotlib.pyplot as plt

%matplotlib inline

plt.rcParams['font.sans-serif']=['SimHei']
plt.rcParams['axes.unicode_minus']=False
```

```python
G = nx.star_graph(7)

nx.draw(G, with_labels=True)
```

### 计算PageRank节点重要度

```python
pagerank = nx.pagerank(G, alpha=0.8)

pagerank
```
