### 彩色图像

```python
# 1. 下载图像样本
!wget https://www.dropbox.com/s/l98leemr7r5stnm/Hemanvi.jpeg
```

```python
# 2. 导入相关程序包并加载图像
import cv2, matplotlib.pyplot as plt
%matplotlib inline
img = cv2.imread('Hemanvi.jpeg')
```

```python
# 3. 裁剪图像
img = img[50:250,40:240,:]
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
# 4. 绘制获得的图像
plt.imshow(img)
print(img.shape)
```

```python
# 5. 可以如下得到右下角的3 x 3 像素数组:
crop = img[-3:, -3:]
```

```python
# 6. 输出并绘制像素值
print(crop)
plt.imshow(crop)
```
